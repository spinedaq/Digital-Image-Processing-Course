#Interfaz Orientada  a Objetos:
Si bien es fácil generar gráficos rápidamente con el módulo matplotlib.pyplot , se recomienda el uso del enfoque orientado a objetos, ya que brinda más control y personalización de sus gráficos. La mayoría de las funciones también están disponibles en la clase matplotlib.axes.Axes paquete matplotlib, módulo axes y clase Axes.

La idea principal detrás del uso del método orientado a objetos más formal es crear objetos de figura y luego simplemente llamar a métodos o atributos fuera de ese objeto. Este enfoque ayuda a manejar mejor un lienzo que tiene varios gráficos.

En la interfaz orientada a objetos, Pyplot se usa solo para algunas funciones, como la creación de figuras, y el usuario crea y realiza un seguimiento explícito de los objetos de figura y ejes. En este nivel, el usuario utiliza Pyplot para crear figuras y, a través de esas figuras, se pueden crear uno o más objetos de ejes. Estos objetos de ejes se utilizan para la mayoría de las acciones de trazado.

Para empezar, creamos una instancia de figura que proporciona un lienzo vacío.

```python
fig = plt.figure() #Instancia de figura(objeto figura). Osea que figure es una clase para crear objetos figura.
```

#Subplots Function()

```python
import matplotlib.pyplot as plt
fig,a =  plt.subplots(2,2) #fig será el objeto figura y a será una tupla con las posibles posiciones de las figuras en el lienzo(en este caso 2 filas, 2 columnas). Aquí se crea el objeto figure y el axes en cascad y queda en a
import numpy as np
x = np.arange(1,5)
a[0][0].plot(x,x*x) #En la posición (0,0)
a[0][0].set_title('square')#titulo de la fifura de la posición (0,0)
a[0][1].plot(x,np.sqrt(x)) #En la posición (0,1)
a[0][1].set_title('square root') #titulo de la fifura de la posición (0,1)
a[1][0].plot(x,np.exp(x)) #En la posición (1,0)
a[1][0].set_title('exp') #titulo de la fifura de la posición (1,0)
a[1][1].plot(x,np.log10(x)) #En la posición (1,1)
a[1][1].set_title('log') #titulo de la fifura de la posición (1,1)
plt.show()
```

#Función subplot2grid():

```python
import matplotlib.pyplot as plt
a1 = plt.subplot2grid((3,3),(0,0),colspan = 2) #objeto axes(cuadricula 3x3, posición (0,0), rango de columna: 2)
a2 = plt.subplot2grid((3,3),(0,2), rowspan = 3) #objeto axes(cuadricula 3x3, posición (0,2), rango de fila: 3)
a3 = plt.subplot2grid((3,3),(1,0),rowspan = 2, colspan = 2) #objeto axes(cuadricula 3x3, posición (1,0), rango de columna:2 y rango de fila: 2)  Subplot2grid es otro constructor de la clase axes y no se aplica sobre figure
import numpy as np
x = np.arange(1,10)
a2.plot(x, x*x) #con las carácteristicas delobjeto 2 grafico
a2.set_title('square')
a1.plot(x, np.exp(x)) #con las carácteristicas delobjeto 1 grafico
a1.set_title('exp')
a3.plot(x, np.log(x)) #con las carácteristicas delobjeto 1 grafico
a3.set_title('log')
plt.tight_layout() #Ajuste del relleno entre y alrededor de las gráficas
plt.show()
```

#Formating Axes(Formato de Ejes):
A veces, uno o unos pocos puntos son mucho más grandes que la mayor parte de los datos. En tal caso, la escala de un eje debe establecerse como logarítmica en lugar de la escala normal. Esta es la escala logarítmica. En Matplotlib, es posible establecer la propiedad xscale o vscale del objeto de ejes en 'registro'.

A veces también se requiere mostrar alguna distancia adicional entre los números de eje y la etiqueta del eje. La propiedad del rotulador de cualquiera de los ejes (x o y o ambos) se puede establecer en el valor deseado.

Ambas características anteriores se demuestran con la ayuda del siguiente ejemplo. La subparcela de la derecha tiene una escala logarítmica y la de la izquierda tiene su eje x con una etiqueta a más distancia.

```python
import matplotlib.pyplot as plt
import numpy as np
fig, axes = plt.subplots(1, 2, figsize=(10,4))
x = np.arange(1,5)
#2 gráficos en la pocisión 0:
axes[0].plot( x, np.exp(x))
axes[0].plot(x,x**2)
axes[0].set_title("Normal scale")
#2 gráficos en la pocisión 1:
axes[1].plot (x, np.exp(x))
axes[1].plot(x, x**2)
axes[1].set_yscale("log") #A la figura de la posición 1  le dy escala logarítmica al eje y
axes[1].set_title("Logarithmic scale (y)")
axes[0].set_xlabel("x axis") #titulo para el eje x de la figura de la pocisión 0
axes[0].set_ylabel("y axis") #titulo para el eje y de la figura de la pocisión 0
axes[0].xaxis.labelpad = 10 #Poniendo la etiqueta del eje x a 10 de distancia
axes[1].set_xlabel("x axis")#titulo para el eje x de la figura de la pocisión 1
axes[1].set_ylabel("y axis") #titulo para el eje x de la figura de la pocisión 1
plt.show()
```

#Ejes gemelos:
Se considera útil tener ejes dobles x o y en una figura. Más aún, al trazar curvas con diferentes unidades juntas. Matplotlib admite esto con las funciones twinx y twiny.



```python
import matplotlib.pyplot as plt
import numpy as np
fig = plt.figure()
a1 = fig.add_axes([0,0,1,1])
x = np.arange(1,11)
a1.plot(x,np.exp(x)) #método graficar(gráfico exponencial)
a1.set_ylabel('exp')
#graficando un ejemelo(para graficar log(X))
a2 = a1.twinx()
a2.plot(x, np.log(x),'ro-')
a2.set_ylabel('log')
fig.legend(labels = ('exp','log'),loc='upper left') #leyenda
plt.show()
```

Al comparar varias cantidades y al cambiar una variable, es posible que deseemos un gráfico de barras en el que tengamos barras de un color para un valor de cantidad.

Podemos trazar múltiples gráficos de barras jugando con el grosor y las posiciones de las barras. La variable de data contiene tres series(listas) de cuatro valores. El siguiente script mostrará tres gráficos de barras de cuatro barras. Las barras tendrán un espesor de 0,25 unidades. Cada gráfico de barras se desplazará 0,25 unidades del anterior. El objeto de datos es un multidict que contiene el número de estudiantes aprobados en tres ramas de una facultad de ingeniería durante los últimos cuatro años.

```python
import numpy as np
import matplotlib.pyplot as plt
data = [[30, 25, 50, 20],
[40, 23, 51, 17],
[35, 22, 45, 19]]
X = np.arange(4)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
#Los 3 graficos de barras:
ax.bar(X + 0.00, data[0], color = 'b', width = 0.25)
ax.bar(X + 0.25, data[1], color = 'g', width = 0.25)
ax.bar(X + 0.50, data[2], color = 'r', width = 0.25)
```

El gráfico de barras apiladas apila barras que representan diferentes grupos una encima de la otra. La altura de la barra resultante muestra el resultado combinado de los grupos.

El parámetro botton opcional de la función pyplot.bar() le permite especificar un valor inicial para una barra. En lugar de correr de cero a un valor, irá desde abajo hasta el valor. La primera llamada a pyplot.bar() traza las barras rojas. La segunda llamada a pyplot.bar() traza las barras azules, con la parte inferior de las barras azules en la parte superior de las barras rojas.

```python
import numpy as np
import matplotlib.pyplot as plt
N = 5
menMeans = (20, 35, 30, 35, 27)
womenMeans = (25, 32, 34, 20, 25)
ind = np.arange(N) # the x locations for the groups
width = 0.35
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(ind, menMeans, width, color='r') #Barras rojas
ax.bar(ind, womenMeans, width,bottom=menMeans, color='b') #Barras azules, que tienen como parte inferior(la parte superior de las barras rojas)
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')

#Etiquetas al eje x:
ax.set_xticks(ind)
ax.set_xticklabels(['G0','G1','G2','G3','G4'])
#Etiquetas al eje y
ax.set_yticks(np.arange(0, 81, 10))
ax.legend(labels=['Men', 'Women']) #Legenda
plt.show()
```

#Gráfico Circular:
El método .pie(x, labels, colors, Autopct) me permite graficar gráficos circulares, donde:
x: Espacio muestral

labels: Es una lista que proporciona las cadenas para cada porción del circulo

Colors(opcional):Secuencia de colores

autopcto: cadena, utilizada para etiquetar las cuñas con su valor numérico. La etiqueta se colocará dentro de la cuña. La cadena de formato será fmt%pct.


```python
from matplotlib import pyplot as plt
import numpy as np
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
langs = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [23,17,35,29,12]
ax.pie(students, labels = langs,autopct='%1.2f%%')
plt.show()
```

#Diagrama de Dispersión:


Los diagramas de dispersión se utilizan para trazar puntos de datos en los ejes horizontal y vertical en un intento de mostrar cuánto afecta una variable a otra. Cada fila en la tabla de datos está representada por un marcador, la posición depende de sus valores en las columnas establecidas en los ejes X e Y. Se puede configurar una tercera variable para que se corresponda con el color o el tamaño de los marcadores, agregando así otra dimensión al gráfico.

#Gráfico de contorno:

Los gráficos de contorno (a veces llamados gráficos de nivel) son una forma de mostrar una superficie tridimensional en un plano bidimensional. Grafica dos variables predictoras XY en el eje y y una variable de respuesta Z como contornos. Estos contornos a veces se denominan cortes z o valores de respuesta iso.

Una gráfica de contorno es adecuada si desea ver cómo cambia el valor Z en función de dos entradas X e Y, de modo que Z = f(X,Y). Una línea de contorno o isolínea de una función de dos variables es una curva a lo largo de la cual la función tiene un valor constante.

Las variables independientes x e y generalmente están restringidas a una cuadrícula regular llamada meshgrid. El numpy.meshgrid crea una cuadrícula rectangular a partir de una matriz de valores x y una matriz de valores y.

La API de Matplotlib contiene funciones de contorno() y contornof() que dibujan líneas de contorno y contornos rellenos, respectivamente. Ambas funciones necesitan tres parámetros x, y y z.

```python
import numpy as np
import matplotlib.pyplot as plt
xlist = np.linspace(-3.0, 3.0, 100)
ylist = np.linspace(-3.0, 3.0, 100)
X, Y = np.meshgrid(xlist, ylist) #Devuelve una matrix de coordanadas a partir de los vectores coordenadas x Y y
Z = np.sqrt(X**2 + Y**2)
fig,ax=plt.subplots(1,1)
cp = ax.contourf(X, Y, Z) #Diagrama de controno
fig.colorbar(cp) # Add a colorbar to a plot
ax.set_title('Filled Contours Plot')
ax.set_xlabel('x (cm)')
ax.set_ylabel('y (cm)')
plt.show()
```

#Diagarama de Aljaba:

Un diagrama de carcaj muestra los vectores de velocidad como flechas con componentes (u, v) en los puntos (x, y).

el método .quiver(x,y,u,v) raza vectores como flechas en las coordenadas especificadas en cada par de elementos correspondiente en x e y. Donde:

X:	Array 1D o 2D, secuencia. Las coordenadas x de las ubicaciones de las flechas

y:	Array 1D o 2D, secuencia. Las coordenadas y de las ubicaciones de las flechas

u:	Array 1D o 2D, secuencia. Las componentes x de los vectores flecha

v:	Array 1D o 2D, secuencia. Las componentes y de los vectores flecha

C(opcional):	Array 1D o 2D, secuencia. Los colores de la flecha

```python
#Ejemplo de diagrama de aljaba simple:
import matplotlib.pyplot as plt
import numpy as np
x,y = np.meshgrid(np.arange(-2, 2, .2), np.arange(-2, 2, .25))
z = x*np.exp(-x**2 - y**2)
v, u = np.gradient(z, .2, .2) #Gradiente de z
fig, ax = plt.subplots()
q = ax.quiver(x,y,u,v) #diagrama de aljaba
plt.show()
```

Al igual que los diagramas de caja, los diagramas de violín se utilizan para representar la comparación de una distribución variable (o distribución de muestra) entre diferentes "categorías".

#Gráficos Tridimiensionales:

Aunque Matplotlib se diseñó inicialmente teniendo en cuenta solo el trazado bidimensional, algunas utilidades de trazado tridimensional se crearon sobre la pantalla bidimensional de Matplotlib en versiones posteriores, para proporcionar un conjunto de herramientas para la visualización de datos tridimensionales. Los gráficos tridimensionales se habilitan al importar el kit de herramientas mplot3d , incluido con el paquete Matplotlib.

Se pueden crear ejes tridimensionales pasando la palabra clave projection='3d' a cualquiera de las rutinas normales de creación de ejes.

```python
from mpl_toolkits import mplot3d #Del módulo mpl_toolkits importo el método mplt3d, que es el que me permite construir gráficos en 3D
import numpy as np
import matplotlib.pyplot as plt
fig = plt.figure()
ax = plt.axes(projection='3d') #Objeto ax peretenciente a la clase axes(veamos que este es otro constructor de la clase axes)(projection='3d', es la palabra clave para que la máquina entienda que es un gráfico 3d)
z = np.linspace(0, 1, 100)
x = z * np.sin(20 * z)
y = z * np.cos(20 * z)
ax.plot3D(x, y, z, 'gray') #Gráfico en 3D(x,y,z) con este método
ax.set_title('3D line plot')
plt.show()
```

#Gráfico de Controno 3D:
La función ax.contour3D() crea un gráfico de contorno tridimensional. Requiere que todos los datos de entrada estén en forma de cuadrículas regulares bidimensionales, con los datos Z evaluados en cada punto. Aquí, mostraremos un diagrama de contorno tridimensional de una función sinusoidal tridimensional.

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
def f(x, y):
   return np.sin(np.sqrt(x ** 2 + y ** 2))
	
x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y) #Función sinusoidal tridimiensional

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary') #Grafico el gráfico tridimensional, con escala de colores binary(blanco y negro)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3D contour')
plt.show()
```

#Gráfico de Estructura Aalámbrica:

El diagrama de estructura alámbrica toma una cuadrícula de valores y la proyecta sobre la superficie tridimensional especificada, y puede hacer que las formas tridimensionales resultantes sean muy fáciles de visualizar. La función plot_wireframe() se usa para el propósito

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
def f(x, y):
   return np.sin(np.sqrt(x ** 2 + y ** 2))
	
x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y) #Función sinusoidal tridimensional

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_wireframe(X, Y, Z, color='black') #Gráfico de estrutura inalámbrica(color negro)
ax.set_title('wireframe')
plt.show()
```

#Gráfico de Superficie 3D:

El gráfico de superficie muestra una relación funcional entre una variable dependiente designada (Y) y dos variables independientes (X y Z). El gráfico es un gráfico complementario del gráfico de contorno. Un gráfico de superficie es como un gráfico de estructura alámbrica, pero cada cara de la estructura alámbrica es un polígono relleno. Esto puede ayudar a la percepción de la topología de la superficie que se está visualizando. La función plot_surface() x, y y z como argumentos.

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
x = np.outer(np.linspace(-2, 2, 30), np.ones(30)) #Calcula el producto exterior de los 2 vectores argumentos
y = x.copy().T # transpose #En y guardo una copia de x transpuesta
z = np.cos(x ** 2 + y ** 2)

fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_surface(x, y, z,cmap='viridis', edgecolor='none') #Gráfico de superficie 3D
ax.set_title('Surface plot')
plt.show()
```

#Trabajar con  Texto:

Matplotlib tiene una amplia compatibilidad con texto, incluida la compatibilidad con expresiones matemáticas, compatibilidad con TrueType para salidas vectoriales y ráster, texto separado por nueva línea con rotaciones arbitrarias y compatibilidad con Unicode. Matplotlib incluye su propio matplotlib.font_manager que implementa un algoritmo de búsqueda de fuentes compatible con W3C multiplataforma.

El usuario tiene un gran control sobre las propiedades del texto (tamaño de fuente, grosor de fuente, ubicación y color del texto, etc.). Matplotlib implementa una gran cantidad de comandos y símbolos matemáticos de TeX.

La siguiente lista de comandos se utiliza para crear texto en la interfaz de Pyplot:

text:Agregue texto en una ubicación arbitraria de los ejes.

annotate:Agregue una anotación, con una flecha opcional, en una ubicación arbitraria de los Ejes.

xlabel:Agregue una etiqueta al eje x de los ejes.

ylabel:Agregue una etiqueta al eje y de los ejes.

title:Agregue un título a los Ejes(al gráfico).

figtext	:Agregue texto en una ubicación arbitraria de la figura.

suptitle:Agregue un subtítulo a la figura.

Todas estas funciones crean y devuelven una instancia de matplotlib.text.Text() .

Los siguientes scripts demuestran el uso de algunas de las funciones anteriores:

```python
import matplotlib.pyplot as plt
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.set_title('axes title')
ax.set_xlabel('xlabel')
ax.set_ylabel('ylabel')
ax.text(3, 8, 'boxed italics text in data coords', style='italic', bbox = {'facecolor': 'red'}) #que el texto inicie en la coordenada (3,8), stilo de letra 'iytalic' y color del fondo rojo
ax.text(2, 6, r'an equation: $E = mc^2$', fontsize = 15) #que el texto inicie en la coordenada (2,6), tamaño de letra=15
ax.text(4, 0.05, 'colored text in axes coords',verticalalignment = 'bottom', color = 'green', fontsize = 15)#que el texto inicie en la coordenada (4,0.05), alineamiento vertical, color de letra verde y tamaño de letra 15
ax.plot([2], [1], 'o') #grafica el punto (2,1) como una bola
ax.annotate('annotate', xy = (2, 1), xytext = (3, 4),arrowprops = dict(facecolor = 'black', shrink = 0.05)) #una plecah de la coordenada 3,4 hacia la coordenada 2,1(que en la coordenada de la cabeza ponga 'annotate) y de color negro
ax.axis([0, 10, 0, 10])
plt.show()
```
